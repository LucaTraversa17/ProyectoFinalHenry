```python
from google.colab import auth
auth.authenticate_user()

# bibliotecas
from google.cloud import storage
import pandas as pd
import io

# cliente de almacenamiento
client = storage.Client()

# bucket y archivo
bucket_name = 'pf_datawarehouse'
file_name = 'metadata_google_restaurantes.json'

# obtener el bucket
bucket = client.get_bucket(bucket_name)

# obtener el blob (archivo)
blob = bucket.blob(file_name)

# descargar el contenido del archivo como string
file_content = blob.download_as_string()

# leer el contenido como DataFrame
df = pd.read_json(io.BytesIO(file_content))

# mostrar el DataFrame
df.head()
```

```python
df.isnull().sum()
```

```python
df_new = df.drop(columns=['description','price','relative_results','url','state','hours'])
df_new.head()
```

```python
# Desempaquetar la columna 'data'
df_expanded = pd.json_normalize(df_new['MISC'])

# Combinar el dataframe original con el dataframe expandido
df_final = df_new.drop(columns=['MISC']).join(df_expanded)

```

```python
df_final.head()
```

```python
df_final.isnull().sum()
```

Los valores de caracteristicas son la mayoría nulos, procedemos a eliminarlos.


```python
df_new = df_new.drop(columns = ['MISC'])
df_new.head()
```

```python
# prompt: remove [ ] from column category

import re

df_new['category'] = df_new['category'].apply(lambda x: re.sub(r'\[|\]', '', x))

df_new.head()
```

```python
df_new = df_new.drop_duplicates()
```

```python
# prompt: exportar la tabla al bucket etl_archivos

# definir el nombre del archivo a exportar
file_name = 'MetaDataGoogleRestaurants.csv'

# convertir el DataFrame a CSV
csv_data = df_new.to_csv(index=False)

# crear un nuevo blob en el bucket de destino
blob = client.bucket('etl_archivos').blob(file_name)

# subir el contenido del DataFrame al blob
blob.upload_from_string(csv_data)

# imprimir un mensaje de confirmación
print(f"El archivo '{file_name}' ha sido exportado al bucket 'etl_archivos'.")
```
